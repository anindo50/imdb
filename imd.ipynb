```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
import pandas as pd
myurl = 'https://www.imdb.com/list/ls051981806/'
open = ureq(myurl)
pagehtml = open.read()
open.close()
pagesoup = soup(pagehtml, "lxml")

names = []
years = []
matr = []
rating = []
runtime = []
typ = []

y=0
x = pagesoup.find_all('h3')
for y in range(0,50):
    for i in x[y].find_all('a'):
        y = y+1
        names.append(i.text)
```

```python
y=0
x = pagesoup.find_all('h3')
for y in range(0,50):
    for i in x[y].find_all('span',class_='lister-item-year text-muted unbold'):
        y = y+1
        years.append(i.text.replace(u",", u" "))
```

```python
y=0
contain = pagesoup.find_all(class_='lister-item mode-detail')
for y in range(0,50):
    for container in contain[y].find_all('span',class_='metascore favorable'):
        y = y + 1
        matr.append(container.text) 
```

```python
x = pagesoup.find_all(class_='ipl-rating-star small')
for y in range(0,50):
    for i in x[y].find(class_='ipl-rating-star__rating'):
        rating.append(i)
```

```python
x = pagesoup.find_all(class_='ratings-metascore')
for y in range(0,25):
    for i in x[y].find_all('span',class_='metascore unfavorable'):
        matr.append(i.text)
```

```python
for i in pagesoup.find_all(class_='runtime'):
    runtime.append(i.text)
```

```python
for i in pagesoup.find_all(class_='genre'):
    typ.append(i.text.replace('\n','').replace(',',''))
```

```python
df = pd.DataFrame({'Movie':names})
df1 = pd.DataFrame({'Year':years})
df2 = pd.DataFrame({'Rating':rating})
df3 = pd.DataFrame({'Metascore':matr})
df4 = pd.DataFrame({'Runtime':runtime})
df5 = pd.DataFrame({'Type':typ})
ff = pd.concat([df,df1,df2,df3,df4,df5],axis=1)
ff.to_csv('imdb.csv',encoding = 'utf-8',header=True)
ff
```
